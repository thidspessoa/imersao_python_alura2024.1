```python
# Formatação personalizada para valores em reais (R$)
def formatar_reais(valor):
    # Converter o valor para o formato de moeda
    valor_formatado = "R${:,.2f}".format(valor)
    # Substituir o ponto por vírgula
    valor_formatado = valor_formatado.replace('.', ',')
    # Separar o valor em partes
    partes = valor_formatado.split(',')
    # Verificar se há mais de uma parte (ou seja, casas decimais)
    if len(partes) > 1:
        # Substituir vírgula por ponto apenas na última parte (casas decimais)
        partes[-1] = partes[-1].replace(',', '.')
    # Juntar as partes novamente
    valor_formatado = '.'.join(partes)
    # Retornar o valor formatado
    return valor_formatado
```

```python
df_analise_saldo['variacao_rs_formatado'] = df_analise_saldo['variacao_rs'].map(formatar_reais)
```

*para criar um grafico, criamos uma variavel chamada fig e atribuimos a ela a chamada da bilbioteca de criação de graficos, pelo apelido q definimos. Chamamos a função bar pois queremos criar um gráfico em barra (você irá selecionar a função de acordo com o tipo de gráfico que deseja criar).*

```python
df_analise_faixa_idade = df_principal_subiu.groupby("Cat_idade")["variacao_rs"].sum().reset_index()
df_analise_faixa_idade
```

```python
total_empresas_por_faixa_etaria = df_principal.groupby('Cat_idade').size()
df_analise_faixa_idade["quantidade_empresas"] = total_empresas_por_faixa_etaria.values
df_analise_faixa_idade['variacao_rs_formatado'] = df_analise_faixa_idade['variacao_rs'].map(formatar_reais)
df_analise_faixa_idade = df_analise_faixa_idade.drop(columns=["variacao_rs"])
df_analise_faixa_idade
```

Ao utilizar .values, estamos extraindo os valores do objeto Series resultante da operação groupby().size(). Isso é necessário porque queremos adicionar uma nova coluna ao DataFrame df_empresas, e precisamos garantir que os índices dos valores correspondam aos índices do DataFrame original.

Quando realizamos operações de agregação como groupby().size(), o pandas retorna um objeto Series com os valores resultantes da operação de contagem e seus índices são os valores únicos da coluna agrupada. Se tentarmos adicionar diretamente essa série como uma nova coluna em outro DataFrame, o pandas tentará alinhar os índices e isso pode não funcionar como esperado.

Portanto, ao utilizar .values, estamos convertendo a Series para um array NumPy unidimensional, o que nos fornece apenas os valores, sem os índices. Dessa forma, podemos ter certeza de que estamos adicionando os valores corretos à nova coluna em df_empresas, mantendo a ordem correta dos dados.

Em resumo, .values nos dá acesso apenas aos valores da série, permitindo-nos adicionar esses valores como uma nova coluna ao DataFrame sem se preocupar com os índices.

```python
fig = px.pie(df_analise_faixa_idade, values="quantidade_empresas", names="Cat_idade")
fig.show()
```
